```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("20241002175215.csv")
df[['total_duration', 'load_duration', 'prompt_eval_duration']] /= 1e9


# Plotting total_duration, load_duration, and prompt_eval_duration
fig, ax = plt.subplots()

df.plot(kind='bar', x='model', y=['total_duration', 'load_duration', 'prompt_eval_duration'], ax=ax)
plt.ylabel('Time in seconds')
plt.title('Duration for Different Models')
plt.show()

# Plotting eval_count and eval_duration
fig, ax = plt.subplots()
df.plot(kind='bar', x='model', y=['eval_count','prompt_eval_count'], ax=ax)
plt.title('Evaluation Count and Evaluation Duration for Different Models')
plt.show()
```
